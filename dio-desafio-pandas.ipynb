# Visualização de Dados


```python
      fig, ax = plt.subplots(1,1)
      fig.set_size_inches(5,5)
      df.loc['Brazil'].plot(kind='pie', ax=ax, autopct='%1.1f%%')
      ax.set_ylabel('')
      ax.set_xlabel('Brazil')
```

```python
for ind in range(10):
      fig, ax = plt.subplots(1,1)
      fig.set_size_inches(5,5)
      df.iloc[ind].plot(kind='pie', ax=ax, autopct='%1.1f%%')
      ax.set_ylabel('')
      ax.set_xlabel(df.index[ind])
```
